```python
# Yelp api
import requests
import json
from pprint import pprint
from yelp.client import Client
# from yelp_api_key import yelp_client_id
# from yelp_api_key import yelp_api_key
from urllib.parse import quote
API_KEY = api_key
API_HOST = 'https://api.yelp.com'
SEARCH_PATH = '/v3/businesses/search'
BUSINESS_PATH = '/v3/businesses/'  

DEFAULT_TERM = 'supermarket'
DEFAULT_LOCATION = '1734 E 41st St, Los Angeles, CA 90058'
# DEFAULT_DISTANCE = 5
SEARCH_LIMIT = 50

def request(host, path, api_key, url_params=None):
    
    url_params = url_params or {}
    url = '{0}{1}'.format(host, quote(path.encode('utf8')))
    headers = {
        'Authorization': 'Bearer %s' % api_key,
    }
    print(u'Querying {0} ...'.format(url))
    response = requests.request('GET', url, headers=headers, params=url_params)
    return response.json()

response = request(host=API_HOST,
                  path=SEARCH_PATH,
                  api_key=API_KEY,
                  url_params={"location":DEFAULT_LOCATION, "term":DEFAULT_TERM, "limit": SEARCH_LIMIT})

pprint(response, indent=4)
```

```python
home_coords = [34.007832,-118.242081]

home_address =  '1734 E 41st St, Los Angeles, CA 90058'

from  geopy.geocoders import Nominatim
geolocator = Nominatim()

def radius_around_foodbank(foodbank, miles):
    loc = geolocator.geocode(foodbank)
    five_miles_north_of_home = loc.latitude - (miles/60)
    five_miles_south_of_home = loc.latitude + (miles/60)
    five_miles_west_of_home = loc.longitude - (miles/60)
    five_miles_east_of_home = loc.longitude + (miles/60)
    return five_miles_south_of_home, five_miles_north_of_home, five_miles_west_of_home, five_miles_east_of_home  
# print(five_miles_south_of_home)
radius_around_foodbank(home_address, 5)[0]
```

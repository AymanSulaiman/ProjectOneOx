```python
lat_foodbank =loc_food_bank.latitude
radians_lat_foodbank = np.radians(lat_foodbank)

lon_foodbank =loc_food_bank.longitude
radians_lon_foodbank = np.radians(lon_foodbank)

earth_radius = 3958.8

radius_miles = 3.75

rad_lat = [lat_foodbank +(((np.cos(x))*radius_miles/1)/60) for x in np.arange(0,2*np.pi,0.1)]
rad_lon = [lon_foodbank +(((np.sin(x))*radius_miles/1)/60) for x in np.arange(0,2*np.pi,0.1)]

min_lat = radians_lat_foodbank - np.radians((2*radius_miles)/earth_radius)
max_lat = radians_lat_foodbank + np.radians((2*radius_miles)/earth_radius)

min_lon = radians_lon_foodbank - np.radians((2*radius_miles)/earth_radius/np.cos(radians_lon_foodbank))
max_lon = radians_lon_foodbank + np.radians((2*radius_miles)/earth_radius/np.cos(radians_lon_foodbank))


supermarkets_filtered_v1 = supermarket_df[
                                          (supermarket_df['Latitude']>=min(rad_lat)) 
                                        & (supermarket_df['Latitude']<=max(rad_lat))
                                        & (supermarket_df['Longitude']<=max(rad_lon))
                                        & (supermarket_df['Longitude']>=min(rad_lon))
                                            ]

supermarkets_filtered_v1.to_csv('supermarkets_filtered_5_mile_radius.csv', index=False)
```
